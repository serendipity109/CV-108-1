```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('Lenna.png')
# convert the image into grayscale
#gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = img[:,:,2]
print(gray)
plt.imshow(gray,'gray')
```

```python
blur = cv2.GaussianBlur(gray, (7,7), 9)
```

```python
plt.imshow(blur, 'gray')
```

```python
# Sobel filtering
Fx = np.array([[[-1,0,1], [-2,0,2], [-1,0,1]]])
Fx = np.squeeze(Fx/8)
Fy = np.array([[[-1,-2,-1], [0,0,0], [1,2,1]]])
Fy = np.squeeze(Fy/8)
```

```python
def conv(image, mask):
  # Height and width of output image
    Hout = image.shape[0] - mask.shape[0] + 1
    Wout = image.shape[1] - mask.shape[1] + 1
    output = np.zeros(shape = (Hout, Wout))
    for i in range(output.shape[0]):
        for j in range(output.shape[1]):
            output[i, j] = np.multiply(image[i : i + mask.shape[0], j : j + mask.shape[1]], mask).sum()
    return output
```

```python
a = conv(blur, Fx)
b = conv(blur, Fy)
```

```python
c = np.sqrt(np.add(np.square(a), np.square(b)))
c = c / c.max() * 255
```

```python
plt.imshow(c, 'gray')
```

```python
theta = np.arctan2(b, a)
```

```python
def non_max_suppression(img, D):
    M, N = img.shape
    Z = np.zeros((M,N), dtype=np.int32)
    angle = D * 180. / np.pi
    angle[angle < 0] += 180

    
    for i in range(1,M-1):
        for j in range(1,N-1):
            try:
                q = 255
                r = 255
                
               #angle 0
                if (0 <= angle[i,j] < 22.5) or (157.5 <= angle[i,j] <= 180):
                    q = img[i, j+1]
                    r = img[i, j-1]
                #angle 45
                elif (22.5 <= angle[i,j] < 67.5):
                    q = img[i+1, j-1]
                    r = img[i-1, j+1]
                #angle 90
                elif (67.5 <= angle[i,j] < 112.5):
                    q = img[i+1, j]
                    r = img[i-1, j]
                #angle 135
                elif (112.5 <= angle[i,j] < 157.5):
                    q = img[i-1, j-1]
                    r = img[i+1, j+1]

                if (img[i,j] >= q) and (img[i,j] >= r):
                    Z[i,j] = img[i,j]
                else:
                    Z[i,j] = 0

            except IndexError as e:
                pass
    
    return Z
```

```python
z = non_max_suppression(c, theta)
```

```python
plt.imshow(z, 'gray')
```

```python
def threshold(img, lowThresholdRatio, highThresholdRatio):
    
    highThreshold = img.max() * highThresholdRatio;
    lowThreshold = highThreshold * lowThresholdRatio;
    
    M, N = img.shape
    res = np.zeros((M,N), dtype=np.int32)
    
    weak = np.int32(25)
    strong = np.int32(255)
    
    strong_i, strong_j = np.where(img >= highThreshold)
    zeros_i, zeros_j = np.where(img < lowThreshold)
    
    weak_i, weak_j = np.where((img <= highThreshold) & (img >= lowThreshold))
    
    res[strong_i, strong_j] = strong
    res[weak_i, weak_j] = weak
    
    return (res, weak, strong)
```

```python
ziv, siv, hd = threshold(z, 0.1, 0.3)
```

```python
plt.imshow(ziv, 'gray')
```

```python
def hysteresis(img, weak, strong=255):
    M, N = img.shape  
    for i in range(1, M-1):
        for j in range(1, N-1):
            if (img[i,j] == weak):
                try:
                    if ((img[i+1, j-1] == strong) or (img[i+1, j] == strong) or (img[i+1, j+1] == strong)
                        or (img[i, j-1] == strong) or (img[i, j+1] == strong)
                        or (img[i-1, j-1] == strong) or (img[i-1, j] == strong) or (img[i-1, j+1] == strong)):
                        img[i, j] = strong
                    else:
                        img[i, j] = 0
                except IndexError as e:
                    pass
    return img
```

```python
outcome = hysteresis(ziv, siv)
```

```python
plt.imshow(outcome, 'gray')
```

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('Lenna.png', 0)
print('Original Dimensions : ',img.shape)

def downsampling(img):
    scale_percent = 50 # percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    return cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

py256 = downsampling(img)
py128 = downsampling(py256)
py64 = downsampling(py128)
py32 = downsampling(py64)
```

```python
def upsampling(img):
    scale_percent = 200 # percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    return cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

error512 = img - upsampling(py256)
error256 = py256 - upsampling(py128)
error128 = py128 - upsampling(py64)
error64 = py64 - upsampling(py32)
```

```python
def reconstruct(py32, error64, error128, error256, error512):
    t1 = upsampling(py32) + error64
    t2 = upsampling(t1) + error128
    t3 = upsampling(t2) + error256
    t4 = upsampling(t3) + error512
    plt.imshow(t4, 'gray')
    return t4
```

```python
q = reconstruct(py32, error64, error128, error256, error512)
```
